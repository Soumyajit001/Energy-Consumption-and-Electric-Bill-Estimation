```python
def sliding_windows(data, seq_length):
    x = []
    y = []
    for i in range(len(data) - seq_length - 1):
        _x = data[i : (i+seq_length)]
        _y = data[i+seq_length]
        x.append(_x)
        y.append(_y)
    return np.array(x), np.array(y)

sc = MinMaxScaler()
training_data = sc.fit_transform(training_set)

seq_length = 4
x, y = sliding_windows(training_data, seq_length)

train_size = int(len(y) * 0.67)
test_size = len(y) - train_size

data_x = Variable(torch.Tensor(np.array(x)))
data_y = Variable(torch.Tensor(np.array(y)))

train_x = Variable(torch.Tensor(np.array(x[0:train_size])))
train_y = Variable(torch.Tensor(np.array(y[0:train_size])))

test_x = Variable(torch.Tensor(np.array(x[train_size:len(x)])))
test_y = Variable(torch.Tensor(np.array(x[train_size:len(y)])))


'''
def sliding_windows(data, seq_length):: This line defines a function named sliding_windows that takes two arguments: data, which is the input time series data, and seq_length, which is the length of the input sequences.

x = [] and y = []: These lines initialize empty lists to store input sequences (x) and corresponding output values (y).

for i in range(len(data) - seq_length - 1):: This line iterates over the indices of the input data, excluding the last seq_length + 1 elements. This ensures that there are enough elements remaining in the data for both input and output sequences.

_x = data[i : (i+seq_length)]: This line extracts a subsequence of length seq_length from the input data, starting from index i, and assigns it to the variable _x.

_y = data[i+seq_length]: This line extracts the element immediately following the _x subsequence and assigns it to the variable _y. This element will be the output value corresponding to the input sequence _x.

x.append(_x) and y.append(_y): These lines append the input sequence _x and output value _y to the respective lists x and y.

return np.array(x), np.array(y): Finally, this line converts the lists x and y into NumPy arrays and returns them as a tuple. The input sequences are returned as x, and the corresponding output values are returned as y.

sc = MinMaxScaler(): This line creates an instance of the MinMaxScaler class from the sklearn.preprocessing module. This scaler is used to scale the input data to a specified range (usually between 0 and 1).

training_data = sc.fit_transform(training_set): This line applies the Min-Max scaling transformation to the training_set data. The fit_transform method fits the scaler to the data (training_set) and then transforms it. The scaled data is stored in the training_data variable.

seq_length = 4: This line sets the length of the input sequences to 4. This means that each input sequence will contain 4 consecutive elements from the time series data.

x, y = sliding_windows(training_data, seq_length): This line calls the sliding_windows function with the scaled training_data and the specified seq_length. It generates input-output pairs (x for input sequences and y for corresponding output values) using the sliding window approach.

train_size = int(len(y) * 0.67): This line calculates the size of the training dataset as approximately 67% of the total number of input-output pairs (y). It converts the result to an integer using the int function.

test_size = len(y) - train_size: This line calculates the size of the testing dataset as the remaining portion after subtracting the training size from the total number of input-output pairs (y).

data_x = Variable(torch.Tensor(np.array(x))) and data_y = Variable(torch.Tensor(np.array(y))): These lines convert the input sequences (x) and output values (y) into PyTorch tensors and wrap them in Variable objects. This step is not necessary in PyTorch versions after 0.4.0, as Variable is deprecated.

train_x = Variable(torch.Tensor(np.array(x[0:train_size]))) and train_y = Variable(torch.Tensor(np.array(y[0:train_size]))): These lines extract the first train_size input-output pairs from the full dataset (x and y) to create the training set (train_x and train_y).

test_x = Variable(torch.Tensor(np.array(x[train_size:len(x)]))) and test_y = Variable(torch.Tensor(np.array(x[train_size:len(y)]))): These lines extract the remaining input-output pairs after the training set to create the testing set (test_x and test_y). There's a mistake here where x is used instead of y. It should be corrected to test_y = Variable(torch.Tensor(np.array(y[train_size:len(y)]))).
'''
```

# LSTM CLASS CODE

```python
class LSTM(nn.Module):

    def __init__(self, num_classes, input_size, hidden_size, num_layers):
        super(LSTM, self).__init__()
        
        self.num_classes = num_classes
        self.num_layers = num_layers
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.seq_length = seq_length
        
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size,
                            num_layers=num_layers, batch_first=True)
        
        self.fc = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        h_0 = Variable(torch.zeros(
            self.num_layers, x.size(0), self.hidden_size))
        
        c_0 = Variable(torch.zeros(
            self.num_layers, x.size(0), self.hidden_size))
        
        # Propagate input through LSTM
        ula, (h_out, _) = self.lstm(x, (h_0, c_0))
        
        h_out = h_out.view(-1, self.hidden_size)
        
        out = self.fc(h_out)
        
        return out
```

# MODEL TRAINING

```python
num_epochs = 10000
learning_rate = 0.03

input_size = 2
hidden_size = 15
num_layers = 1

num_classes = 1

lstm = LSTM(num_classes, input_size, hidden_size, num_layers)

criterion = torch.nn.MSELoss()    # mean-squared error for regression
optimizer = torch.optim.Adam(lstm.parameters(), lr = learning_rate)
#optimizer = torch.optim.SGD(lstm.parameters(), lr=learning_rate)
#t=trainX*1.01

# Train the model
for epoch in range(num_epochs):
    outputs = lstm(train_x * 1.02)
    optimizer.zero_grad()

    loss = criterion(outputs, train_y)

    loss.backward()
    outputs = nn.Dropout(0.5)
    optimizer.step()
    if epoch % 100 == 0:
        print("Epoch: %d, loss: %1.5f" % (epoch, loss.item()))
```

# Evaluating the Model

```python
lstm.eval()
train_predict = lstm(data_x)

data_predict = train_predict.data.numpy()
data_y_plot = data_y.data.numpy()

# data_predict = sc.inverse_transform(data_predict)
# data_y_plot = sc.inverse_transform(data_y_plot)

plt.axvline(x=train_size, c='r', linestyle='--')

plt.plot(data_y_plot, label='actual')
plt.plot(data_predict,label='Predicted')
plt.suptitle('Energy Time-Series Prediction')
plt.legend(loc="upper right")
plt.show()
```
